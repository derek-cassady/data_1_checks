## Import required libraries

```python
import requests
import pandas as pd 
```

## Pull in data from an API

```python
url = 'https://api.census.gov/data/2020/dec/dhc'
variables = ['NAME,P12_002N,P12_003N,P12_004N,P12_005N,P12_006N,P12_007N,P12_008N,P12_009N,P12_010N,P12_011N,P12_012N,P12_013N,P12_014N,P12_015N,P12_016N,P12_017N,P12_018N,P12_019N,P12_020N,P12_021N,P12_022N,P12_023N,P12_024N,P12_025N']
how = 'state:*'
where = 'county:*'
key = '593ff23df30236471e2bd1165031f208ce43d9f4' 

r = requests.request("GET",url,params={
    "get": variables,
    "for": where,
    "in" : how,
    "key": key
})
```

## Read in data from API to DataFrame

```python
df = pd.DataFrame(r.json())
df.drop([0], inplace = True)
```
